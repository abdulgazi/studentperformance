```python
student_per_p = 'student-por.csv'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_per = pd.read_csv(student_per_p, sep=';', decimal='.', header=0, names=['school', 'sex', 'age',
                                                                                  'address', 'famsize',
                                                                                  'pstatus', 'medu', 'fedu',
                                                                                  'mjob', 'fjob', 'reason',
                                                                                  'guardian', 'traveltime',
                                                                                  'studytime', 'failures',
                                                                                  'schoolsup', 'famsup',
                                                                                  'paid', 'activities',
                                                                                  'nursery', 'higher', 'internet',
                                                                                  'romantic', 'famrel', 'freetime',
                                                                                  'goout', 'dalc', 'walc', 'health',
                                                                                  'absences', 'grade1', 'grade2',
                                                                                  'finalgrade'],index_col=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#to visualise the number of occurences of the categorical column values
%matplotlib inline
romantic_count = student_per['romantic'].value_counts()
sns.set(style="darkgrid")
sns.barplot(romantic_count.index, romantic_count.values, alpha=0.9)
plt.title('FD of romantic column')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Romantic', fontsize=12)
plt.show()
```

```python
#to visualise the number of occurences of the categorical column values
romantic_count = student_per['guardian'].value_counts()
sns.set(style="darkgrid")
sns.barplot(romantic_count.index, romantic_count.values, alpha=0.9)
plt.title('FD of guardian column')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Guardian', fontsize=12)
plt.show()
```

```python
#to visualise the number of occurences of the categorical column values
romantic_count = student_per['pstatus'].value_counts()
sns.set(style="darkgrid")
sns.barplot(romantic_count.index, romantic_count.values, alpha=0.9)
plt.title('FD of pstatus column')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Pstatus', fontsize=12)
plt.show()
```

```python
#Explore relationships
```

```python
#Address-Absences: Students who live in regional areas are absent more often
sns.catplot(x="address", y="absences", kind="box", data=student_per)
```

```python
#Pstatus-Famrel: Students whose parents are together will a better family relationship
sns.catplot(x="pstatus", y="famrel", kind="box", data=student_per)
```

```python
#Address-Walc: Students who live in urban areas drink more
sns.catplot(x="address", y="walc", kind="box", data=student_per)
```

```python
#Guardian-Finalgrade: Students whose guardian is 'other' score less
sns.catplot(x="guardian", y="finalgrade", kind="box", data=student_per)
```

```python
#TravelTime-Health: Students who have less traveltime are more healthy
sns.catplot(x="traveltime", y="health", kind="box", data=student_per)
```

```python
#Romantic-Absences: Students who are in a romantic relationship have more absences
sns.catplot(x="romantic", y="absences", kind="box", data=student_per)
```

```python
#Famrel-Walc: Students who score their family relationship lower drink more on weekends
sns.catplot(x="famrel", y="walc", kind="box", data=student_per)
```

```python
#Walc-Finalgrade: Students who drink more on the weekend score less 
sns.catplot(x="walc", y="finalgrade", kind="box", data=student_per)
```

```python
#Health-Absences: Students with poor health have more absences
sns.catplot(x="health", y="absences", kind="box", data=student_per)
```

```python
#Absences-Finalgrade: Students with more absences will have a lower score
sns.catplot(x="absences", y="finalgrade", kind="box", data=student_per)
```
